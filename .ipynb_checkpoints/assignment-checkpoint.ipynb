### Q1. 

**Tokens**: The individual words or symbols that create text data like emails. Natural Language Processing is primarily about analyzing the frequency and co-occurrence of tokens.

I aggregated all the emails into a single vector, and removed the punctuation and very common words (e.g. "the"). Run the below code chunk to open it, and use the Counter class to look at the most common words:

```python
import pickle
with open('data/all_tokens.pickle', 'rb') as file:
    all_tokens = pickle.load(file)
    
from collections import Counter
token_count = Counter(all_tokens)
token_freq = token_count.most_common()
```

Plot a histogram of the occurrences of tokens. 

What do you notice about the frequency of occurrence of different tokens? How does it look?

```python
import matplotlib.pyplot as plt
import numpy as np

counts = np.array(list(token_count.values()))

plt.figure(figsize=(10,6))
plt.hist(counts, bins=30)
plt.xlabel('Number of occurrences of each token')
plt.ylabel('Number of tokens')
plt.title('Histogram of the Occurrences of Tokens')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(counts, bins=30, log=True)
plt.xlabel('Number of occurrences of each token')
plt.ylabel('Number of tokens (log scale)')
plt.title('Histogram of the Occurrences of Tokens (log scale)')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```

### Q2. 

Load `Phishing_clean.parquet`. This is the text from the e-mails broken into the most common 2,711 tokens and one-hot-encoded as features/covariates. So each row is an e-mail, the `Email Type` takes the value 1 if it's a scam and 0 otherwise, and every other column is a word or symbol that occurs in at least 15 e-mails.

1. Perform an 80/20 train-test split of the data.
2. Run a regression of $y$ on the one-hot-encoded emails. What is the $R^2$ on the test set? On the training set?
3. What words have the largest coefficients in absolute value and most strongly influence predictions? 

**1.**

```python
import pandas as pd

df = pd.read_parquet('./data/Phishing_clean.parquet')
print(df.shape)
df.head()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns=['Email Type'])
y = df['Email Type']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0
)
```

**2.**

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)

# compute R^2
r2_train = model.score(X_train, y_train)
r2_test = model.score(X_test, y_test)

print("R on training set: ", r2_train)
print("R on test set: ", r2_test)
```

**3.**

```python
coef_df = pd.DataFrame({
    'token': X.columns,
    'coef': model.coef_
})
coef_df['abs_coef'] = coef_df['coef'].abs()

top_tokens = coef_df.sort_values('abs_coef', ascending=False).head(10)

print("\nTop 10 words with largest coefficients:")
print(top_tokens)
```

### Q3. 

Take the matrix of one-hot-encoded tokens (the data, less the outcome variable, `Email Type`) and perform a principal components analysis decomposition with two components. Plot the first two principal components in a scatter plot, and hue the points by whether they are a phishing scam or not. Do you notice any patterns?

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca = pca.fit(X)
X_pca = pca.transform(X)

print('Their explained variance ratio: ', pca.explained_variance_ratio_)

# scatter plot
plt.figure(figsize=(8,6))
plt.scatter(
    X_pca[y == 0, 0], X_pca[y == 0, 1],
    alpha=0.6, s=20, label="Otherwise (0)"
)
plt.scatter(
    X_pca[y == 1, 0], X_pca[y == 1, 1],
    alpha=0.6, s=20, label="Scam (1)"
)
plt.xlabel(f"Principal Component 1 ({pca.explained_variance_ratio_[0]*100:.2f}% var)")
plt.ylabel(f"Principal Component 2 ({pca.explained_variance_ratio_[1]*100:.2f}% var)")
plt.title("PCA of One-Hot-Encoded Emails (2 Components)")
plt.legend()
plt.tight_layout()
plt.show()
```

In the above scatter plot, it can be seen that PCA of the one-hot-encoded emails onto the first two principal components have substantial overlap between phishing and legitimate emails. Based on the explained variance ratio, the first two components explain only 2% of the total variance, indicating that the distinguishing features between email types is multi-dmensional. While there are a few clusters with higher concentrations of scam emails, there is no clear linear separation in two dimensions.

### Q4.

Run a linear regression of $y$ on the first 2,610 principal components of $X$. What is the $R^2$ on the training and test sets? (I used cross validation to determine that 2,610 was approximately optimal, but not all 2,711 components.)

How does this performance compare to the linear regression?

```python
pca = PCA(n_components=2610)
pca = pca.fit(X_train)
X_train_pca = pca.transform(X_train)
X_test_pca  = pca.transform(X_test)

print("Total variance ratio: ", pca.explained_variance_ratio_.sum())

pca_lr = LinearRegression()
pca_lr.fit(X_train_pca, y_train)

r2_train_pca = pca_lr.score(X_train_pca, y_train)
r2_test_pca  = pca_lr.score(X_test_pca, y_test)

print("R on training set: ", r2_train_pca)
print("R on test set: ", r2_test_pca)
```

The linear regression model trained directly on the one-hot-encoded token matrix produced an extremely negative R-squared value on the test set. This indicates numerical instability from hgihly correlated binary features. After applying PCA with 2,610 components, the model achieved a train R-squared value of 0.768 and a test R-squared value of 0.658, demonstrating that dimensionality reduction stabilized the regression. Additionally, the total variance ratio of 0.9998 shows that predictive performance was preserved with nearly all relevant variation captured.

### Q5.

Explain briefly in your own words what the advantage is in using the principal components to run this high-dimensional regression, rather than the original data.

Using principal components to run the high-dimensional regression reduces the dimensionality of the data while retaining most of its variance. This helps eliminate redundance among highly correlated features and makes the regression model more stable and less prone to overfitting or numerical issues. Overall, the model becomes more efficient and generalizes better to unseen data without sacrificing a lot of predictive accuracy.
